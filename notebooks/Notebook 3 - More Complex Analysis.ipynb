# Merging data

```python
employee_dataframe = pd.DataFrame([{"employee":"Tim","department":"finance"},{"employee":"Bob","department":"finance"},{"employee":"Jack","department":"HR"},{"employee":"Jim","department":"logistics"},{"employee":"Alice"}])
```

```python
employee_dataframe
```

```python
department_dataframe = pd.DataFrame([{"department":"finance","reports to":"CFO"},{"department":"HR","reports to":"COO"},{"department":"logistics","reports to":"COO"},{"department":"operations","reports to":"COO"}])
```

```python
department_dataframe
```

# Joining strategy matters (how!) :
* inner
* outer
* left
* right

```python
employee_dataframe.merge(department_dataframe,left_on="department",right_on="department",how="inner")
```

```python
employee_dataframe.merge(department_dataframe,left_on="department",right_on="department",how="outer")
```

```python
employee_dataframe.merge(department_dataframe,left_on="department",right_on="department",how="left")
```

```python
employee_dataframe.merge(department_dataframe,left_on="department",right_on="department",how="right")
```
